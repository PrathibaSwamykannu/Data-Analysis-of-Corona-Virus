**Data Cleaning**

```python
'''Data Cleaning'''
#Update the datatype of Last Update and Date to datetime
data['Last Update'] = data['Last Update'].apply(pd.to_datetime)
data['Date'] = data['Date'].apply(pd.to_datetime)
```

```python
#Remove the S.No column as there is no real use for it in the analysis
data.drop(['Sno'],axis=1,inplace=True)
data.head()
```

We can see now that the S.No column has been dropped

```python
#check the datatype of Last Update and Date
data.info()
```

The datatype of both the columns have been changed to datetime for convenience

**Data Visualisation with Map**

```python
world_d = dict(
   name=list(data['Country']),
    lat=[-25.27,12.57,56.13,61.92,46.23,51.17,22.32,20.59,41.87,36.2,22.2,35.86,4.21,28.39,12.87,1.35,35.91,7.87,23.7,15.87,37.09,23.42,14.06,],
   lon=[133.78,104.99,-106.35,25.75,2.21,10.45,114.17,78.96,12.56,138.25,113.54,104.19,101.98,84.12,121.77,103.82,127.77,80.77,120.96,100.99,-95.71,53.84,108.28],
   Confirmed=list(data['Recovered'])
)
world_data = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in world_d.items() ]))
world_data = world_data.fillna(method='ffill')

fig = go.Figure()
fig.add_trace(go.Scattergeo(
        lat=world_data['lat'],
        lon=world_data['lon'],
        mode='markers',
        marker=dict(
            size=17,
            color='rgb(255, 0, 0)',
            opacity=0.7
        ),
        text=world_data['name'],
        hoverinfo='text'
    ))

fig.add_trace(go.Scattergeo(
        lat=world_data['lat'],
        lon=world_data['lon'],
        mode='markers',
        marker=dict(
            size=8,
            color='rgb(242, 177, 172)',
            opacity=0.7
        ),
        hoverinfo='none'
    ))

fig.update_layout(
        autosize=True,
        hovermode='closest',
        showlegend=False,
        title_text = '<b>Countries with reported confirmed cases, Deaths, Recovered </b>',
        font=dict(family="Arial, Balto, Courier New, Droid Sans", color='black'),
    geo = go.layout.Geo(
        showframe = False,
        showcoastlines = True,
        showcountries = True,
        landcolor = "rgb(229, 229, 229)",
        countrycolor = "black",
        coastlinecolor = "black",
        projection_type="natural earth"
    ))

 
fig.show()
```
